## 2. Now write a function that returns the largest element on a list

```python
#your code here
def greatest(lst):
    biggest = lst[0]       # assuming the first element as the biggest
    
    for num in lst:        # we loop through the list
        if num > biggest:  # if we find a bigger number
            biggest = num  # we update the biggest number
            
    return biggest
```

```python
# This will test your function 
test_greatest(greatest)
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
#your code here
    result = 1
    i = 1
    
    while i <= n:
        result = result * i
        i = i + 1
    

    return result
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

```python
# This will test your function 
test_factorial(factorial)
```

## 7. Write a function that takes a list and returns a list of the unique values.

`NOTE: You cannot use set. 🤔`

```python
def unique(lst_un):
#your code here
    result = []

    for item in lst_un:
        found = False

        for existing in result:
            if item == existing:
                found = True

        if found == False:
            result.append(item)

    return result
    
```

```python
# This will test your function 
test_unique(unique)
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... 🧐`

```python
def mode_counter(arr):
#your code here
    max_count = 0
    mode = None

    for item in arr:
        count = 0

        for element in arr:
            if element == item:
                count = count + 1

        if count > max_count:
            max_count = count
            mode = item

    return mode
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. 😉`

```python
def st_dev(list_sd):
#your code here
    # Step 1: calculate mean
    total = 0
    count = 0

    for num in list_sd:
        total = total + num
        count = count + 1

    mean = total / count

    # Step 2: calculate variance
    variance_sum = 0

    for num in list_sd:
        diff = num - mean
        variance_sum = variance_sum + (diff * diff)

    variance = variance_sum / (count - 1) # Tried using count instead of count - 1, but it was not correct. I had to use count - 1 to get the correct answer.

    # Step 3: square root (power of 0.5)
    std_dev = variance ** 0.5

    return std_dev
```

```python
# This will test your function 
test_stdev(st_dev)
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
#your code here
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    
    for letter in alphabet:
        found = False
        
        for char in string:
            if char == letter or char == letter.upper():
                found = True
        
        if found == False:
            return False

    return True
```

```python
# This will test your function 
test_pangram(pangram)
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! 🤪`

```python
def sort_alpha(string):
#your code here
    words = []
    current_word = ""

    # Step 1: manually split by commas
    for char in string:
        if char == ",":
            words.append(current_word)
            current_word = ""
        else:
            current_word = current_word + char

    # add last word
    words.append(current_word)

    # Step 2: sort
    words = sorted(words)

    # Step 3: manually join with commas
    result = ""
    first = True

    for word in words:
        if first == True:
            result = word
            first = False
        else:
            result = result + "," + word

    return result
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
#your code here
    has_lower = False
    has_upper = False
    has_digit = False
    has_special = False
    length = 0

    for char in password:
        length = length + 1

        # check lowercase
        if char >= 'a' and char <= 'z':
            has_lower = True

        # check uppercase
        if char >= 'A' and char <= 'Z':
            has_upper = True

        # check digit
        if char >= '0' and char <= '9':
            has_digit = True

        # check special character
        if not ((char >= 'a' and char <= 'z') or 
                (char >= 'A' and char <= 'Z') or 
                (char >= '0' and char <= '9')):
            has_special = True

    # final check
    if length >= 8 and has_lower and has_upper and has_digit and has_special:
        return True
    else:
        return False
```

```python
# This will test your function 
test_pass(check_pass)
```
